```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
import pandas as pd
path = "/content/drive/MyDrive/newpro/threeclass_problem/bert_changedlevels.csv"
dataset = pd.read_csv(path)
dataset.head()
```

```python
def PreProcessing(dataset):
  #Remove First row (Indexing from Excel)
  dataset = dataset.iloc[:, 1:] 
  
  #Shuffling
  dataset = dataset.sample(frac = 1) #frac denotes the fraction of samples to be returned. 

  # distributing the dataset into two components X and Y
  X = dataset.iloc[:, 0:-1].values
  y = dataset.iloc[:, -1].values

  #Scaling (Standardization)
  sc = StandardScaler()
  X = sc.fit_transform(X)

  print(X.shape)
  print(y.shape)
  return X, y
```

```python
X, y = PreProcessing(dataset)
```

PCA Function

```python
def PCA_fn(X, name, i): 
    pca = PCA(n_components = i)
    X = pca.fit_transform(X)
    explained_variance = pca.explained_variance_ratio_
    print(X.shape)

    #Store result as csv
    temp = name+"_"+str(i)+"_features.csv"
    df1 = pd.DataFrame(X)
    df1.to_csv(temp) 
    files.download(temp)
```

```python
arr = [50, 100, 150, 200, 250]
for i in arr:
  PCA_fn(X, "bert_withstopwordchangelevels", i)
```


Implement Above Code For All Word Embeddings

```python
for i in arr:
  #BERT
  path = "/content/drive/MyDrive/newpro/threeclass_problem/fasttext_changedlevels.csv"
  d1 = pd.read_csv(path)
  X, y = PreProcessing(d1)
  PCA_fn(X, "fasttext_withstopwordchangelevels", i) 
  print("___________________________________________________________________________________________________") 
```
