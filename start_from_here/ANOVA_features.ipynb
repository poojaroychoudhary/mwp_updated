```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
path = "/content/drive/MyDrive/newpro/threeclass_problem/tfidf_changedlevels.csv"
dataset = pd.read_csv(path)
dataset.head()
```

```python
def PreProcessing(dataset):
  #Remove First row (Indexing from Excel)
  dataset = dataset.iloc[:, 1:] 
    
  #Shuffling
  dataset = dataset.sample(frac = 1) #frac denotes the fraction of samples to be returned. 

  # distributing the dataset into two components X and Y
  X = dataset.iloc[:, 0:-1].values
  y = dataset.iloc[:, -1].values

  #Scaling (Standardization)
  sc = StandardScaler()
  X = sc.fit_transform(X)

  print(X.shape)
  print(y.shape)
  return X, y
```

```python
dataset.drop(dataset.columns[[0]], axis=1, inplace=True)
dataset.head()
```

```python
X, y = PreProcessing(dataset)
```

```python
def anova_fn(X, y, name, i): 
    fvalue_selector = SelectKBest(f_classif, k = i)
    X_best = fvalue_selector.fit_transform(X, y)
    print(X_best.shape)

    #Store result as csv
    temp = name+"_"+str(i)+"_features.csv"
    df1 = pd.DataFrame(X_best)
    df1.to_csv(temp) 
    files.download(temp)
```

```python
arr = [50, 100, 150, 200, 250]
for i in arr:
  anova_fn(X, y, "tfidf_withstopwordchangelevels", i)
```

```python
for i in arr:
  #Glove
  path1 = "/content/drive/MyDrive/newpro/threeclass_problem/glove_changedlevels.csv"
  dataset1 = pd.read_csv(path1)
  X, y = PreProcessing(dataset1)
  anova_fn(X, y, "glove_withstopwordchangelevels", i) 
  print("___________________________________________________________________________________________________")
```
