```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
import pandas as pd
path = "/content/drive/MyDrive/newpro/threeclass_problem/bert_changedlevels.csv"
dataset = pd.read_csv(path)
dataset.head() 
```

```python
dataset.drop(dataset.columns[[0]], axis=1, inplace=True)
dataset.head()
```

```python
def PreProcessing(dataset):
  #Remove First row (Indexing from Excel)
  dataset = dataset.iloc[:, 1:] 
    
  #Shuffling
  dataset = dataset.sample(frac = 1) #frac denotes the fraction of samples to be returned. 

  # distributing the dataset into two components X and Y
  X = dataset.iloc[:, 0:-1].values
  y = dataset.iloc[:, -1].values

  #Scaling (Standardization)
  sc = StandardScaler()
  X = sc.fit_transform(X)

  print(X.shape)
  print(y.shape)
  return X, y
```

```python
X, y = PreProcessing(dataset)
```

```python
def LDA_fn(X, y, name):
  lda = LinearDiscriminantAnalysis(n_components = 2)
  lda.fit(X, y)
  transformed = lda.transform(X)
  print(transformed.shape)

  #Store result as csv
  temp = name+"_"+"2_features.csv"
  df1 = pd.DataFrame(transformed)
  df1.to_csv(temp) 
  files.download(temp)
```

```python
LDA_fn(X, y, "bert_2lda_withstopwordchangelevels")
```

```python
path = "/content/drive/MyDrive/newpro/threeclass_problem/fasttext_changedlevels.csv"
dataset = pd.read_csv(path)
X, y = PreProcessing(dataset)
LDA_fn(X, y, "fasttext_2lda_withstopwordchangelevels")
print("___________________________________________________________________________________________________")
```
